```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
a=pd.read_csv('technical_support_data-2.csv')
a
```

```python
a=a.drop('PROBLEM_TYPE',axis=1)
a
```

```python
from scipy.stats import zscore
```

```python
a=a.apply(zscore)
a
```

```python
sns.pairplot(a)
```

```python
from sklearn.cluster import KMeans
```

```python
# elbow method
k_rang=range(1,10)
sse=[]
for k in k_rang:
    km=KMeans(n_clusters=k)
    km.fit(a)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rang,sse)
```

```python
# k=3 and k=6

# for k=3
km=KMeans(n_clusters=6)
km.fit(a)
```

```python
y_predicted=km.fit_predict(a)
y_predicted
```

```python
a['cluster']=y_predicted
a
```

```python
a.corr()
```

```python
a=a.drop(['Avg_pending_calls','Avg_resol_time','In_warranty_percent'],axis=1)
a
```

```python
sns.pairplot(a)
```
